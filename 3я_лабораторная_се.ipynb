Двустороннее отображение: $x_{n+1}=\frac{λx_n}{1+x_n^3}, 0<λ<500$

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.axes as axs
import math
```

```python
def func(lamb):
  def next_x(xn):
    return (lamb*xn)/(1+xn**3)
  return next_x
```

```python
# Функция для построения бифуркационной диаграммы
def bifurcation_diagram(l_range, x0_range, num_iterations, burn_in):
    plt.figure(figsize=(9, 7))
    for l in l_range:
        for x0 in x0_range:
            x = x0
            for _ in range(burn_in):
              f=func(l)
              x = f(x)
            x_values = []
            for _ in range(num_iterations):
                f=func(l)
                x = f(x)
                x_values.append(x)

            plt.plot([l] * len(x_values), x_values, ',', color='blue', alpha=0.25)

    plt.title('Бифуркационная диаграмма для отображения')
    plt.xlabel('Параметр $\lambda$')
    plt.ylabel('x')
    plt.show()
```

```python
x0_range = np.linspace(-0.9, 0.9, 100)

lamb_bifurcation = np.linspace(0.1, 25, 100)
bifurcation_diagram(lamb_bifurcation, x0_range, num_iterations=500, burn_in=1000)
```
